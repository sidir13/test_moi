# Présentation d'ElevenLabs

## Qui est ElevenLabs ?

**ElevenLabs** est une entreprise spécialisée dans l’**IA générative audio**, fondée en **2022** à Londres par deux amis d’enfance polonais.    
Les fondateurs sont **Piotr Dąbkowski** (CTO, ex-Google DeepMind) et **Mati Staniszewski** (CEO, ex-Palantir).    

Ils se sont rencontrés au lycée Copernicus de Varsovie et partagent une passion commune pour la technologie et l’innovation. 

## Origine et mission

L’idée d’ElevenLabs vient de la frustration de regarder des films américains mal doublés en polonais.    
Les fondateurs voulaient créer une technologie capable de reproduire des voix humaines de manière ultra-réaliste pour améliorer le doublage et la narration.    

La **mission** d’ElevenLabs est de construire la plateforme audio-IA la plus complète au monde et de rendre toute information accessible dans n’importe quelle voix, langue et son.    
L’entreprise cherche à rendre la communication avec la technologie aussi naturelle à l’oral qu’à l’écrit.  

## Croissance et positionnement

ElevenLabs adopte une approche de **research-first company**, en plaçant la recherche en IA au centre de sa stratégie.   
Depuis le lancement public en 2023, la plateforme a rapidement attiré des millions d’utilisateurs et de nombreux clients entreprises.    

En quelques années, ElevenLabs est devenue une **licorne** valorisée autour de 3,3 milliards de dollars.   
Elle est aujourd’hui utilisée par de grandes entreprises médias, de création de contenu, de jeux vidéo et de tech.  

## Principales capacités audio IA

ElevenLabs propose une suite de produits audio basés sur l’IA :  

- **Text-to-Speech (TTS)** : conversion de texte en voix naturelle, avec gestion des émotions et de la prosodie.
- **Speech-to-Text (STT)** : transcription audio multilingue (modèle Scribe et évolutions). 
- **Voice Cloning** : clonage de voix (instantané ou professionnel) à partir d’échantillons audio. 
- **Voice Design** : création de voix originales à partir de descriptions textuelles. 
- **Dubbing** : doublage multilingue en conservant le timbre et le style de la voix d’origine. 
- **Text-to-Dialogue** : génération de dialogues multi-locuteurs naturels. 
- **Voice Remixing, sound effects et musique** : transformation de voix, génération d’effets sonores et de musique via l’IA.  

La **Voice Library** contient plus de 3 000 voix créées par la communauté et réutilisables dans différents projets. 

## Modèles TTS phares

ElevenLabs propose plusieurs modèles de synthèse vocale, chacun optimisé pour un cas d’usage : 

- **Eleven v3** : modèle très expressif, orienté émotion et performance, supportant plus de 70 langues (limite ~5 000 caractères). 
- **Multilingual v2** : optimisé pour les textes longs (audiobooks, podcasts), 29 langues, limite ~10 000 caractères. 
- **Flash v2.5** : très faible latence (~75 ms hors réseau), 32 langues, jusqu’à 40 000 caractères, coût réduit. 
- **Turbo v2.5** : compromis qualité/vitesse avec faible latence (~250–300 ms) et coût abaissé. 

## Dimension technologique et éthique

Les modèles d’ElevenLabs reposent sur des architectures de deep learning avancées capables de générer des voix quasi indiscernables de voix humaines.    
Ils gèrent des éléments subtils comme les pauses, les intonations, les rires ou les hésitations, ce qui les rend adaptés aux usages narratifs et conversationnels.   

L’entreprise met aussi l’accent sur la **sécurité** et l’**usage responsable** de l’IA vocale, avec des mécanismes de détection, du watermarking audio et des garde-fous contre le clonage abusif.

# Text-to-Speech avec l’API ElevenLabs

L’API Text-to-Speech (TTS) d’ElevenLabs permet de convertir du texte en audio avec une prosodie naturelle, une gestion des émotions et un support multilingue (32+ langues selon le modèle choisi).    
Elle s’appuie sur différents modèles (Eleven v3, Multilingual v2, Flash v2.5, Turbo v2.5) qui font varier le compromis entre qualité, latence, longueur maximale de texte et coût.    

Dans les exemples suivants, nous verrons comment :
- Appeler l’API TTS pour générer un fichier audio à partir d’un texte.
- Choisir un modèle et une voix.
- Configurer le format de sortie (mp3, PCM, μ-law, A-law, Opus).  
- Gérer la génération temps réel (streaming) dans les cas d’usage interactifs.  

## Configuration de base (Python + clé API)

Pour utiliser l’API ElevenLabs en Python, on configure d’abord la clé API (en variable d’environnement ou directement dans le code).  
Ici, nous allons utiliser la bibliothèque officielle `elevenlabs` (si disponible) ou faire un appel HTTP classique si besoin.

```python
!pip install elevenlabs --quiet

import os
from elevenlabs import ElevenLabs

# À adapter : renseigner ta clé API ELEVENLABS_API_KEY (variable d'environnement)
api_key = os.getenv("ELEVENLABS_API_KEY", "TON_API_KEY_ICI")

client = ElevenLabs(api_key=api_key)
```

## Génération simple : texte → MP3

Ici, on envoie un texte à l’API TTS et on récupère un flux audio encodé en MP3 (format par défaut).    
On spécifie :
- le modèle (ex. `eleven_multilingual_v2`, `eleven_flash_v2_5`, etc.),  
- la voix (ID d’une voix existante ou générée dans la Voice Library),  
- éventuellement le format de sortie si on souhaite autre chose que du MP3 (PCM, μ-law, A-law, Opus).  


```python
# Exemple minimal de génération TTS et sauvegarde dans un fichier MP3

text = "Bonjour, ceci est un exemple de synthèse vocale générée avec ElevenLabs."
voice_id = "EXAMPLE_VOICE_ID"  # Remplace par l'ID d'une voix de ta librairie

# Appel type (adapté au client officiel actuel)
audio = client.text_to_speech.convert(
    voice_id=voice_id,
    model_id="eleven_multilingual_v2",  # ou "eleven_flash_v2_5", "eleven_turbo_v2_5", etc.  
    text=text,
    output_format="mp3_44100_128"       # MP3 44.1 kHz, 128 kbps (voir section formats)  
)

# Sauvegarde sur disque
with open("output_simple_tts.mp3", "wb") as f:
    f.write(audio)
```

## Choix du modèle TTS

Les principaux modèles TTS d’ElevenLabs ont des profils différents :  

- **Eleven v3** : modèle le plus expressif émotionnellement, support ~70 langues, limite de 5 000 caractères, bon pour des dialogues riches.   
- **Multilingual v2** : voix naturelles, 29 langues, 10 000 caractères, très stable pour les contenus longs (audiobooks, podcasts).   
- **Flash v2.5** : ultra faible latence (~75 ms hors latence réseau), 32 langues, 40 000 caractères, coût réduit, idéal pour des scénarios temps réel.   
- **Turbo v2.5** : compromis entre qualité et vitesse, faible latence (~250–300 ms) et prix réduit.   

Dans le code, ce choix se fait via le paramètre `model_id`.


```python
texts = {
    "eleven_multilingual_v2": "Ceci est un test avec le modèle Multilingual v2.",
    "eleven_flash_v2_5": "Ceci est un test avec le modèle Flash v2.5, optimisé pour la faible latence.",
    "eleven_turbo_v2_5": "Ceci est un test avec le modèle Turbo v2.5."
}

for model_id, txt in texts.items():
    audio = client.text_to_speech.convert(
        voice_id=voice_id,
        model_id=model_id,
        text=txt,
        output_format="mp3_44100_128"
    )
    filename = f"tts_{model_id}.mp3"
    with open(filename, "wb") as f:
        f.write(audio)
    print(f"Généré : {filename}")
```

## Formats audio disponibles

Le format par défaut de réponse de l’API TTS est **MP3**, mais d’autres formats sont possibles :  

- **MP3** : fréquence 22,05 kHz ou 44,1 kHz, bitrates de 32 à 192 kbps (ex. `mp3_22050_32`, `mp3_44100_128`).  
- **PCM (S16LE)** : 16 bits, fréquences de 16 à 44,1 kHz (ex. `pcm_16000`, `pcm_44100`).  
- **μ-law** : 8 kHz, optimisé pour la téléphonie (ex. `ulaw_8000`).  
- **A-law** : 8 kHz, également orienté téléphonie (ex. `alaw_8000`).  
- **Opus** : 48 kHz, bitrates de 32 à 192 kbps, adapté au streaming (ex. `opus_48000_96`).  

Les formats de meilleure qualité sont disponibles sur les offres payantes (voir la page de pricing API).  


```python
text = "Test de différents formats audio générés par l’API ElevenLabs."

formats = [
    "mp3_44100_128",
    "pcm_16000",
    "ulaw_8000",
    "alaw_8000",
    "opus_48000_96",
]

for fmt in formats:
    audio = client.text_to_speech.convert(
        voice_id=voice_id,
        model_id="eleven_multilingual_v2",
        text=text,
        output_format=fmt
    )
    ext = fmt.split("_")[0]  # mp3, pcm, ulaw, alaw, opus
    filename = f"tts_format_{ext}.bin" if ext != "mp3" else f"tts_format_{ext}.mp3"
    with open(filename, "wb") as f:
        f.write(audio)
    print(f"Généré : {filename} (format interne : {fmt})")
```

## Cas temps réel : modèle Flash v2.5

Pour les applications temps réel (agents vocaux, assistants conversationnels, jeux vidéo, etc.), ElevenLabs propose le modèle **Flash v2.5**, optimisé pour une latence très faible (~75 ms hors application et réseau).     

L’idée est de :
- Utiliser `eleven_flash_v2_5` comme `model_id`. 
- Générer des segments audio courts, ou utiliser les endpoints temps réel/streaming décrits dans la section développeurs (WebSocket ou streaming HTTP).  


```python
import time

segments = [
    "Bonjour,",
    " ceci est une phrase générée en plusieurs segments,",
    " pour simuler un comportement temps réel avec le modèle Flash.",
]

for i, seg in enumerate(segments, start=1):
    audio = client.text_to_speech.convert(
        voice_id=voice_id,
        model_id="eleven_flash_v2_5",
        text=seg,
        output_format="mp3_44100_128"
    )
    filename = f"tts_flash_segment_{i}.mp3"
    with open(filename, "wb") as f:
        f.write(audio)
    print(f"Généré : {filename}")
    time.sleep(0.3)  # simple attente pour imiter un flux séquentiel
```

## Bonnes pratiques TTS

Pour des textes longs (audiobooks, chapitres, scripts), il est conseillé : 
- De découper le texte en segments cohérents (phrases, paragraphes) pour garder une prosodie stable.
- D’utiliser un modèle adapté au long format (Multilingual v2 pour la stabilité, Turbo v2.5 ou Flash v2.5 pour réduire les coûts/latence).
- De surveiller la limite de caractères par requête (par exemple 5 000 pour Eleven v3, 10 000 pour Multilingual v2, 40 000 pour Flash/Turbo v2.5). 

Pour les cas nécessitant une synchronisation fine (karaoké, sous-titres), il est possible de coupler la TTS avec la fonctionnalité de **Forced Alignment** qui aligne texte et audio.
